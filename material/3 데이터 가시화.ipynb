##  여러 그래프 그리기

### subplot (state-based)

```python
# 여러 그래프 그리기
# subplot = state-based
# subplots = object-oriented
fig = plt.figure(figsize=(12,6))
plt.subplot(1,2,1) # plt.subplot(121)로 적어도 됨
plt.scatter(bank_df['duration'], bank_df['balance'])
plt.subplot(1,2,2)
plt.scatter(bank_df['age'], bank_df['balance'])
plt.xlabel('Age')
plt.ylabel('Balance')
plt.show()
```

```python
# 만일 subgraph가 4개면, 221, 222, 223, 224로 표기함
fig = plt.figure(figsize=(12,12))
plt.subplot(2,2,1)
plt.scatter(bank_df['duration'], bank_df['balance'])
plt.subplot(2,2,2)
plt.scatter(bank_df['duration'], bank_df['balance'])
plt.subplot(2,2,3)
plt.scatter(bank_df['duration'], bank_df['balance'])
plt.subplot(2,2,4)
plt.scatter(bank_df['duration'], bank_df['balance'])
plt.show()
```

```python
# 여러 그래프 그리기
# subplot = state-based
# subplots = object-oriented
fig = plt.figure(figsize=(12,6))
plt.subplot(121)
plt.scatter(bank_df['duration'], bank_df['balance'])
plt.subplot(122)
plt.scatter(bank_df['age'], bank_df['balance'])
plt.xlabel('Age')
plt.ylabel('Balance')
plt.show()
```

```python
fig = plt.figure(figsize=(12,6))
plt.subplot(2,1,1)
plt.scatter(bank_df['duration'], bank_df['balance'])
plt.subplot(2,1,2)
plt.scatter(bank_df['age'], bank_df['balance'])
plt.xlabel('Age')
plt.ylabel('Balance')
plt.show()
```

### subplots (object-oriented)

```python
# 여러 그래프 그리기 (subplots)
# subplot = state-based
# subplots = object-oriented
fig, ax = plt.subplots(1,2,figsize=(12,6))
fig.suptitle("Two Scatter Plots")
ax[0].scatter(bank_df['duration'], bank_df['balance'])
ax[1].scatter(bank_df['age'], bank_df['balance'])
ax[0].set_title("Duration & Balance")
ax[0].set_xlabel('Duration')
ax[0].set_ylabel('Balance')
ax[1].set_title("Age & Balance")
ax[1].set_xlabel('Age')
ax[1].set_ylabel('Balance')
plt.show()
```

```python
# 여러 그래프 그리기
# subplot = state-based
# subplots = object-oriented
fig, ax = plt.subplots(2,2,figsize=(12,6))
fig.suptitle("Four Scatter Plots")
ax[0,0].scatter(bank_df['duration'], bank_df['balance'])
ax[0,1].scatter(bank_df['age'], bank_df['balance'])
ax[0,0].set_title("Duration & Balance")
ax[0,0].set_xlabel('Duration')
ax[0,0].set_ylabel('Balance')
ax[0,1].set_title("Age & Balance")
ax[0,1].set_xlabel('Age')
ax[0,1].set_ylabel('Balance')

#
ax[1,0].scatter(bank_df['duration'], bank_df['balance'])
ax[1,1].scatter(bank_df['age'], bank_df['balance'])
ax[1,0].set_title("Duration & Balance")
ax[1,0].set_xlabel('Duration')
ax[1,0].set_ylabel('Balance')
ax[1,1].set_title("Age & Balance")
ax[1,1].set_xlabel('Age')
ax[1,1].set_ylabel('Balance')
plt.show()
```

```python
# create a figure and axes

fig, ax = plt.subplots()

# scatter the sepal_length against the sepal_width
ax.scatter(bank_df['duration'], bank_df['balance'])
# set a title and labels
ax.set_title("Duration & Balance")
ax.set_xlabel('Duration')
ax.set_ylabel('Balance')
plt.show()
```

## 산점도 - 색상 적용하기

## Heatmap (상관관계)

```python
## Heatmap
### 색깔별로 높고 낮음을 표시하는 매트릭스

import numpy as np

# get correlation matrix
corr = bank_df[['age','balance','duration']].corr()
fig, ax = plt.subplots() # default 1,1
plt.pcolor(corr) # .pcolor (숫자를 색을 조정)

# set labels
ax.set_xticks(np.arange(0.5,len(corr.columns))) # 0.5, 1.5, 2.5 (steps=1)
ax.set_yticks(np.arange(0.5, len(corr.columns)))
ax.set_xticklabels(corr.columns)
ax.set_yticklabels(corr.columns)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", # 폰트 기울이기
         rotation_mode="anchor")
plt.colorbar()
plt.show()
```
