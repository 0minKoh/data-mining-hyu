## NULL / NA 처리

python에서는 똑같이 취급한다.
NA -> not available
NULL -> not defined

```python
bank_df.isna() # NA이면 True 아니면 False
```

```python
bank_df.isnull() # isna와 동일한 역할
```

```python
bank_df.isna().sum() # 컬럼끼리 합 -> Missing Value를 확인
```

```python
bank_df.isna().sum().sum()
```

```python
import numpy as np
bank_df.iloc[0,0]=np.nan ## NaN - 숫자로 쓸 수 없음 Not a Number
bank_df
```

```python
bank_df.isna().sum().sum()
```

```python
bank_df['age']=bank_df['age'].fillna(bank_df['age'].mean())
bank_df
```

```python
bank_df.isna().sum().sum()
```

```python
# lambda 이름없는 함수
## lambda x: return값 조건문

# .apply(func): 특정 함수를 적용
bank_df['target']=bank_df['deposit'].apply(lambda x: 1 if x=='yes' else 0)
bank_df
```

```python
# .replace(past, future, inplace)
bank_df['target'].replace([0,1],['no','yes'],inplace=True)
```

```python
# .replace({past: future}, inplace)
bank_df['target'].replace({0:'no',1:'yes'},inplace=True)
```

```python
bank_df
```

```python
bank_df['target'].replace({'no':0,'yes':1},inplace=True)
```

```python
bank_df
```

```python
bank_df['age_cat']=bank_df['age'].apply(lambda x: 'Teenager' if x<=19 else ('Adult' if x<=60 else 'Elderly'))
bank_df['age_cat'].value_counts()
```

```python
def get_age_category(age):
    cat=""
    if age<=19:
        cat="Teenager"
    elif age<=60:
        cat="Adult"
    else:
        cat="Elderly"
    return cat
```

```python
bank_df['age_cat']=bank_df['age'].apply(lambda x: get_age_category(x))
bank_df['age_cat'].value_counts()
```
