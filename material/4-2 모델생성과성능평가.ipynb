## 데이터셋 분할

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(sales_df.drop('flag',axis=1),sales_df['flag'],
                                                  test_size=0.2, random_state=123)
```

## 1. 의사결정나무 (분류)

```python
from sklearn.tree import DecisionTreeClassifier

dt_clf = DecisionTreeClassifier(random_state=11)
dt_clf.fit(X_train, y_train)
dt_pred = dt_clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print('예측 정확도: {0:.4f}'.format(accuracy_score(y_test,dt_pred)))
```

## 2. 로지스틱 회귀

```python
# 타겟 변수가 범주형일 때, 분류할 때 사용
from sklearn.linear_model import LogisticRegression

lr_clf = LogisticRegression()
lr_clf.fit(X_train, y_train)
lr_pred = lr_clf.predict(X_test)
```

```python
print('예측 정확도: {0:.4f}'.format(accuracy_score(y_test,lr_pred)))
```

## 3. 랜덤 포레스트 (분류)

```python
from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier(random_state=0)
rf_clf.fit(X_train , y_train)
rf_pred = rf_clf.predict(X_test)
```

```python
print('예측 정확도: {0:.4f}'.format(accuracy_score(y_test,rf_pred)))
```

## 4. MLP 머신러닝 (분류)

```python
from sklearn.neural_network import MLPClassifier # Multi Layered Perceptron

mlp_clf = MLPClassifier(hidden_layer_sizes=(10,10), max_iter=300)
mlp_clf.fit(X_train , y_train)
mlp_pred = mlp_clf.predict(X_test)
```

```python
print('예측 정확도: {0:.4f}'.format(accuracy_score(y_test,mlp_pred)))
```

# 성능 평가

## Confusion Merics (오분류표)

```python
# confusion matrix

from sklearn.metrics import confusion_matrix

# 의사결정나무 오분류표
confusion_matrix(y_test, dt_pred) # Positive의 기준은 우리가 정해야 함!
```

```python
# 로지스틱회귀 오분류표
confusion_matrix(y_test, lr_pred)
```

```python
# 랜덤포레스트 오분류표
confusion_matrix(y_test, rf_pred)
```

```python
from sklearn.metrics import precision_score , recall_score, accuracy_score

accuracy = accuracy_score(y_test,dt_pred) # 정확도
recall = recall_score(y_test,dt_pred) # 재현율
precision = precision_score(y_test, dt_pred) # 정밀도
F1 = 2 * (recall * precision) / (recall + precision)

print("정확도: {0:.4f}. 재현율: {1:.4f}, 정밀도: {2:.4f}, F1: {3:.4f}".format(accuracy, recall, precision, F1))
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score , recall_score, accuracy_score, classification_report

# Confusion Metrix 사용 함수
## 재사용이 가능하도록 함수화
def get_clf_eval(y_test,pred):
    confusion = confusion_matrix(y_test,pred)
    accuracy = accuracy_score(y_test,pred)
    recall = recall_score(y_test,pred)
    precision = precision_score(y_test,pred)
    print('confusion matrix')
    print(confusion)
    print("정확도: {0:.4f}. 재현율: {1:.4f}, 정밀도: {2:.4f}".format(accuracy, recall, precision))
```

```python
print(classification_report(y_test,dt_pred))
```

```python
get_clf_eval(y_test,dt_pred)
```

```python
get_clf_eval(y_test,lr_pred)
```

```python
get_clf_eval(y_test,rf_pred)
```

```python
get_clf_eval(y_test,mlp_pred)
```
