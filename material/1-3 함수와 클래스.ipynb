````python
def list_max(a):
    max_temp=a[0]
    for i in a:
        if i>max_temp:
            max_temp=i
    return max_temp
````

````python
a=[1,4,5,9,6,7]

list_max(a)
````

## 클래스 상속
자식 클래스 (Sub Class) 부모 클래스 (Super Class)를 상속받게 되면
부모 클래스의 속성과 메소드를 재사용할 수 있음

````python
class Rectangle:
    def __init__(self, width=1, height=1):
        self._width = width
        self._height = height

    def setWidth(self,width):
        self._width=width

    def setHeight(self,height):
        self._height=height

    def getWidth(self):
        return self._width

    def getHeight(self):
        return self._height

    def area(self):
        return self._width*self._height

    def perimeter(self):
        return 2*(self._width+self._height)

    def __str__(self):
        return("Width: "+str(self._width)+" Height: "+str(self._height))
````

````python
r1=Rectangle()
print(r1)
r2=Rectangle(2,4)
print(r2)

print(r2.area())
print(r2.perimeter())
print(r2._height)
print(r2._width)
````

````python
class Student:
    def __init__(self,sid, name,department,year=1):
        self.sid=str(sid)
        self.name=name
        self.department=department
        self.year=year
    def changeDepartment(self, department):
        self.department=department
    def getDepartment(self):
        return self.department
    def __str__(self):
        return("Student ID: "+self.sid+" Name: "+self.name+\
               " Department: "+self.department)
````

````python
s1=Student("202401001","Jong Woo Kim","Business")
print(s1)
print(s1.getDepartment())
s1.changeDepartment('Digital Business')
print(s1.getDepartment())
````

````python
class GradStudent(Student):
    def __init__(self,sid, name,department,advisor):
        super().__init__(sid, name,department)
        self.advisor=advisor
    def setAdvisor(self, prof_name):
        self.advisor=prof_name
    def getAdvisor(self):
        return self.advisor
    def __str__(self):
        return("Student ID: "+self.sid+" Name: "+self.name+\
               " Department: "+self.department+" Advisor: "+self.advisor)
````

````python
s2=GradStudent("202405003","Gil Dong Hong","Finance","Chuel Soo Lee")
print(s2)
s2.changeDepartment("Marketing")
print(s2)
````
