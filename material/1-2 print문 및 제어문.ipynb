# 제어문

## if문

````python
n=5
if n%2==0:
    print("%d is even number."%n)
else:
    print("%d is odd number."%n)
````

````python
n=30
if n%3==0 and n%5==0:
    print("%d is divided by 3 and 5"%n)
elif n%3==0:
    print("%d is divide by 3"%n)
elif n%5==0:
    print("%d is divide by 5"%n)
else:
    print("%d is not divided by both 3 and 5"%n)
````

## for, while 문

````python
# range(0, i+1, steps)
## 0부터 i번째까지 step씩 증가하는 range 객체를 반환
print(range(10))
print(range(1,10))
````

````python
sum=0
for i in range(1,11):
    sum=sum+i
print("sum:%d"%sum)
````

````python
sum=0
i=0
while i<11:
    sum=sum+i
    i=i+1
print("sum:%d"%sum)
````

````python
for i in range(1,11):
  for j in range(1,i+1):
    print("*",end="")
  print()
````

````python
for i in range(1,11):
    print("*"*i)
````
