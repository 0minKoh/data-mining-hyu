# matplotlib을 이용하여 다음을 수행하시오. 

```python
import pandas as pd
import matplotlib.pyplot as plt

boston_df = pd.read_csv('asset/boston_housing.csv')
boston_df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(boston_df['medv'])
plt.show()
```

```python
plt.boxplot(boston_df['medv'])
plt.show()
```

```python
# x, height가 있다는 것!
chas = boston_df['chas'].value_counts()
plt.bar(x=chas.index, height=chas.values)
plt.show()
```

```python
# Value Error! (리스트로 줘야함)
data_list = []
chas_index_list = boston_df['chas'].value_counts().index
for chas_index in chas_index_list:
  data_list.append(boston_df.loc[boston_df['chas'] == chas_index, 'medv'])


plt.boxplot(data_list, labels=chas_index_list)
plt.show()
```

```python
plt.scatter(x=boston_df['crim'], y=boston_df['medv'])
plt.show()
```

```python
boston_df['chas'].value_counts()
```

```python
colors = {0: 'r', 1: 'g'}

for idx, chas in enumerate(boston_df['chas'].values):
  plt.scatter(x=boston_df.iloc[idx, :]['crim'], y=boston_df.iloc[idx, :]['medv'], color=colors[chas])

plt.show()
```

```python
plt.plot(boston_df['medv'])
```

8. 전체 변수들에 대한 히트맵을 그리시오.

```python
import numpy as np

corr = boston_df.corr()
plt.pcolor(corr)
plt.xticks(np.arange(0.5, len(corr.columns)), labels=corr.columns)
plt.yticks(np.arange(0.5, len(corr.columns)), labels=corr.columns)
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1)
plt.pcolor(corr)

ax.set_xticks(np.arange(0.5, len(corr.columns)))
ax.set_yticks(np.arange(0.5, len(corr.columns)))
ax.set_xticklabels(corr.columns)
ax.set_yticklabels(corr.columns)

plt.setp(ax.get_xticklabels(), rotation=45, ha='right', rotation_mode='anchor')
plt.colorbar()
plt.show()
```
