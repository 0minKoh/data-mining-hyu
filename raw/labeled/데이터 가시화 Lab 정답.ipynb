```python
import pandas as pd
import os

os.getcwd()
os.chdir(r'C:\Users\USER\kjw\데이터마이닝_학부_24_2\가시화')


boston_df = pd.read_csv('boston_housing.csv')
boston_df
```

```python
import matplotlib.pyplot as plt

plt.hist(boston_df['medv'])
plt.show()
```

```python
# create figure and axis
fig, ax = plt.subplots()
# plot histogram
ax.hist(boston_df['medv'])
# set title and labels
ax.set_title('Boston Housing Price Median Value')
ax.set_xlabel('Housing Price Median Value')
ax.set_ylabel('Frequency')
plt.show()
```

```python
plt.boxplot(boston_df['medv'])
plt.show()
```

```python
data = boston_df['chas'].value_counts() 
plt.bar(data.index,data.values)
plt.show()
```

```python
chas_near = boston_df[boston_df['chas']==1]['medv']
chas_not_near = boston_df[boston_df['chas']==0]['medv']
data=[chas_near,chas_not_near] 
plt.boxplot(data,labels=['Near Charles','Not Near Charles'])
plt.show()
```

```python
# scatter plot

plt.scatter(boston_df['crim'], boston_df['medv'])
plt.show()
```

```python
# create a figure and axes

fig, ax = plt.subplots()

ax.scatter(boston_df['crim'], boston_df['medv'])
# set a title and labels
ax.set_title('Boston Housing Dataset')
ax.set_xlabel('Crime')
ax.set_ylabel('House Price Median Value')
plt.show()
```

```python
# create a figure and axes

colors = {0:'r', 1:'g'}
fig, ax = plt.subplots()

for i in range(len(boston_df['crim'])): 
    ax.scatter(boston_df['crim'][i], boston_df['medv'][i],color=colors[boston_df['chas'][i]])
# set a title and labels
ax.set_title('Boston Housing Dataset')
ax.set_xlabel('Crime')
ax.set_ylabel('House Price Median Value')
plt.show()
```

```python
plt.plot(range(len(boston_df['medv'])), boston_df['medv'])
```

```python
# create a figure and axis 
fig, ax = plt.subplots() 
# count the occurrence of each class 
data = boston_df['chas'].value_counts() 
# get x and y data 
points = data.index 
frequency = data.values 
# create bar chart 
ax.bar(points, frequency) 
# set title and labels 
ax.set_title('Boston Housing Chas Dummy Variable')
ax.set_xlabel('Near Chas (1) or Not (0)')
ax.set_ylabel('Frequency')
```

```python
# get correlation matrix
import numpy as np
corr = boston_df.corr()
fig, ax = plt.subplots()
plt.pcolor(corr)

# set labels
ax.set_xticks(np.arange(0.5, len(corr.columns)))
ax.set_yticks(np.arange(0.5, len(corr.columns)))
ax.set_xticklabels(corr.columns)
ax.set_yticklabels(corr.columns)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

plt.colorbar()
plt.show()
```
