```python
customer_df['IncomeLevel']=customer_df['YearlyIncome'].apply(lambda x: 'Low' if x<50000 else ('Middle' if x<100000 else 'High'))
customer_df['IncomeLevel'].value_counts()
```

```python
def get_income_level(income):
    level=""
    if income<50000:
        level="Low"
    elif income<100000:
        level="Middle"
    else:
        level="High"
    return level
```

```python
customer_df['IncomeLevel']=customer_df['YearlyIncome'].apply(lambda x: get_income_level(x))
customer_df['IncomeLevel'].value_counts()
```
